### DECISION TREE

### Step 1.  Read in Titanic.csv and observe a few samples, some features are categorical, and others are numerical. If some features are missing, fill them in using the average of the same feature of other samples. Take a random 80% samples for training and the rest 20% for test.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt

data = 'Titanic.csv'
df = pd.read_csv(data)
```

```python
# pclass column data was converted to integers: 1st -> 1, 2nd -> 2, 3rd -> 3
# sex column data was converted to integers: male -> 1, female -> 2
# NA values in the age column were all replaced by average value of all other age values
X = df[["pclass","sex", "age", "sibsp"]]
y = df[["survived"]]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
print(x_train.head())
print(f"x_train and y_train shapes are {x_train.shape} and {y_train.shape}, respectively.")
print(f"x_test and y_test shapes are {x_test.shape} and {y_test.shape}, respectively.")
```

#### 

### Step 2. Fit a decision tree model using independent variables ‘pclass + sex + age + sibsp’ and dependent variable ‘survived’. Plot the full tree.

```python
clfr = tree.DecisionTreeClassifier(criterion="entropy")
clfr.fit(x_train, y_train)
plt.figure(figsize=(50,50))
features = x_train.columns
classes = ['Did Not Survive', 'Survived']
tree.plot_tree(clfr, feature_names=features, class_names=classes, filled=True)
print('Original Tree')
```

#### 

### Step 3. Use the GridSearchCV() function to find the best parameter max_leaf_nodes to prune the tree. Plot the pruned tree which shall be smaller than the tree you obtained in Step 2.

```python
parameters = {'max_leaf_nodes': np.arange(2,50)}
grid = GridSearchCV(clfr, parameters, scoring="accuracy")
grid.fit(x_train, y_train)
print(f"Best parameter {grid.best_params_}")
```

```python
clfr2 = tree.DecisionTreeClassifier(**grid.best_params_, criterion="entropy")
clfr2.fit(x_train, y_train)
plt.figure(figsize=(50,50))
features = x_train.columns
classes = ['Did Not Survive', 'Survived']
tree.plot_tree(clfr2, feature_names=features, class_names=classes, filled=True)

print('Pruned Tree')
```

#### 

### Step 4. Report accuracy for the test set

```python
y_pred = clfr2.predict(x_test)
print("Overall Accuracy:", accuracy_score(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(clfr2, x_test, y_test)
print('Confusion Matrix (Test Dataset)')
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
```

```python
print("Percentage of survivors correctly predicted: {:.2f}%".format((tp/(tp+fn))*100))
print("Percentage of fatalities correctly predicted: {:.2f}%".format((tn/(tn+fp))*100))
```

#### 

#### 

### RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

clfr3 = RandomForestClassifier(max_leaf_nodes=grid.best_params_["max_leaf_nodes"], n_estimators=50, criterion="entropy")
clfr3.fit(x_train, y_train.values.ravel())
y_pred2 = clfr3.predict(x_test)
print("Overall Accuracy:", accuracy_score(y_test, y_pred2))
```

```python
ConfusionMatrixDisplay.from_estimator(clfr3, x_test, y_test)
print('Confusion Matrix (Test Dataset)')
tn2, fp2, fn2, tp2 = confusion_matrix(y_test, y_pred2).ravel()
```

```python
print("Percentage of survivors correctly predicted: {:.2f}%".format((tp2/(tp2+fn2))*100))
print("Percentage of fatalities correctly predicted: {:.2f}%".format((tn2/(tn2+fp2))*100))
```

#### 

#### Observation:

Using the Random Forest Classifier, there is an improvement in favor of one outcome but also a slight decrease in the accuracy of the other. The random forest seems to be compensating the accuracy of either outcome (Survived or Did Not Survive) with the other. In general, the accuracy is not significantly varied compared to that of the Decision Tree. One potential reason for this behavior is the number of columns/classes we use to train the algorithm. With only four columns utilized, the set of outcomes of a single tree does not vary much with the different combinations of its branches. It means that the set of outcomes of any version of the decision tree is close to one another. That said, even if we take multiple random trees in the random forest, if their outcomes do not significantly differ, the accuracy will be around a certain radius of magnitude. 

It is important to note that this observed relationship is true only for the set of arguments selected above. Also, the outcomes may have been skewed by the dataset cleanup that has been performed where NA values in the 'age' column were replaced by the overall average value of the column.
