### 1)	Plot the data (i.e., x-axis for the 1st column, y-axis for the 2nd column

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
df = pd.read_csv("data2.txt", sep=",", header=None, names=["x", "y"])
X_unsorted = np.array(df["x"].tolist())
Y_unsorted = np.array(df["y"].tolist())
df = df.sort_values("x")
X = np.array(df["x"].tolist())
Y = np.array(df["y"].tolist())

plt.plot(X,Y, "ro")
plt.xlabel("$x_{raw}$")
plt.ylabel("$y_{raw}$")
plt.title("Raw Data")
```

### 2) Implement the _Normal equation_ to find the coefficients

```python
from numpy.linalg import inv

X_normal = np.stack((np.ones(np.shape(X.flatten())), X.flatten()), axis=1)

w_normal = np.dot(inv(np.dot(np.transpose(X_normal),X_normal)), np.dot(np.transpose(X_normal), Y))
w_normal = w_normal.reshape(w_normal.shape[0],1)
Y_normal = X_normal@w_normal

print(f"y_normal = {float(w_normal[0])} + {float(w_normal[1])}*x")

plt.plot(X, Y, 'ro', label= "raw data")
plt.plot(X, Y_normal, 'b-', label="learned model")
plt.title("Linear Regression via Normal Equation")
plt.xlabel("$x_{raw}$")
plt.ylabel("$y_{raw}, y_{normal}$")
plt.legend(frameon=True, shadow=True, borderpad=2)
```

### 3) Implement the _Gradient Descent method (batch and stochastic modes)_ to find the coefficients

#### a) Split dataset into 80% for training and 20% for testing.

```python
import math

X_train_count = math.ceil(X.size*0.8)

X_train = X_unsorted[:X_train_count]
Y_train = Y_unsorted[:X_train_count]

X_test = X_unsorted[X_train_count:]
Y_test = Y_unsorted[X_train_count:]
```

#### Plot MSE vs. iteration of each mode for both training set and testing set; compare batch and stochastic modes (with discussion) in terms of accuracy (of testing set) and speed of convergence (Determine an appropriate termination condition, e.g., when cost function is less than a threshold, and/or after a given number of iterations.) 

```python
import random

def GradDes(mode, w_init, Xtrain, Ytrain, Xtest, Ytest, poly_deg, delta, step_size, max_iter):  
# mode -> gradient descent mode: 1 for batch, 2 for stochastic
# w_init -> initial guess for column vector w
# Xtrain -> column vector X of the training data set
# Ytrain -> column vector Y of the training data set
# Xtest -> column vector X of the testing data set
# Ytest -> column vector Y of the testing data set
# poly_deg -> degree of polynomial
# delta -> convergence criterion
# step_size -> value of lambda (step size of descent)
# max_iter -> maximum allowable number of iterations
        
        m = len(Xtrain)  # number of elements in the training vector
        m_test = len(X_test)  # number of elements in the testing vector
        
        iter_cnt = 0 # iterations count
        w = w_init.reshape(len(w_init),1) 
        
        # for training set
        X_gd = np.zeros((m, poly_deg+1))  # container for new X values in the training set per iteration
            
        # for testing set
        X_gd_test = np.zeros((m_test, poly_deg+1))  # container for new X values in the testing set per iteration
            
        # edit column numbers & contents of vector X_gd according to polynomial degree
        for i in range(0, poly_deg+1):
            X_gd[:,i] = np.power(Xtrain,i)
            
        # edit column numbers & contents of vector X_gd_test accroding to polynomial degree
        for i in range(0, poly_deg+1):
            X_gd_test[:,i] = np.power(Xtest,i)
        
        # compute initial mse for training set
        mse = 0
        for i in range(0, m):
            mse += (w.T@X_gd[i,:]-Ytrain[i])**2
        mse = (1/m)*mse
        
        # compute initial mse for testing set
        mse_test = 0
        for i in range(0, m_test):
            mse_test += (w.T@X_gd_test[i,:]-Ytest[i])**2
        mse_test = (1/m_test)*mse_test
        
        # initialize array storage of all mse values
        mse_arr = np.array([mse])
        mse_arr_test = np.array([mse_test])
        
        while True:
            
            # for training set
            mse_tmp = 0
            for i in range(0, m):
                #print(w)
                #print(X_gd[i,:])
                #print(w.T@X_gd[i,:])
                mse_tmp += (w.T@X_gd[i,:]-Ytrain[i])**2
            mse = (1/m)*mse_tmp
            
            # for testing set
            mse_tmp_test = 0
            for i in range(0, m_test):
                mse_tmp_test += (w.T@X_gd_test[i,:]-Ytest[i])**2
            mse_test = (1/m_test)*mse_tmp_test
            
            # test for termination
            if (mse <= delta) or (iter_cnt >= max_iter):
                return (w, iter_cnt, mse_arr, mse_arr_test)
            
            # compute the gradient of the cost function
            mse_grad = 0
            if (mode == 1):  # batch mode
                mse_grad  = np.zeros(w.shape)
                for i in range(0, m):
                    mse_grad += (w.T@X_gd[i,:]-Ytrain[i])*X_gd[i,:].reshape(w.shape)
                mse_grad = (2/m)*mse_grad
            elif (mode == 2):  # stochastic mode
                mse_grad  = np.zeros(w.shape)
                index_rand = random.randint(0, m-1)
                mse_grad = (w.T@X_gd[index_rand,:]-Ytrain[index_rand])*X_gd[index_rand,:].reshape(w.shape)
                mse_grad = 2*mse_grad
            else: 
                raise ValueError("Mode not recognized. Select 1 for batch mode or 2 for stochastic mode.")
            
            
            # update w
            w = w - step_size*mse_grad
            
            # update other array containing mse records
            mse_arr = np.append(mse_arr, [mse])
            mse_arr_test = np.append(mse_arr_test, [mse_test])
            
            # update iteration count
            iter_cnt += 1
            
```

```python
# Function input parameters
w_init = np.array([0, 10])  # initial guess array
err_tol = 9.67  # arbitrary error tolerance for termination criterion
step_sz = 0.01  # leanring rate lambda
max_iter = 5000 # desired max iteration count

w, iter_cnt, mse_arr, mse_arr_test = GradDes(mode=1, w_init=w_init, Xtrain=X_train, Ytrain=Y_train, Xtest=X_test, 
                                     Ytest=Y_test, poly_deg=1, delta= err_tol, step_size=step_sz, max_iter=max_iter)
print(f"learned model: y = {float(w[0])} + {float(w[1])}x")
print(f"iteration count: {iter_cnt}")
print(f"final training MSE: {mse_arr[-1]}")
print(f"final testing MSE: {mse_arr_test[-1]}")

iterations = np.arange(0, iter_cnt+1, 1)

plt.plot(iterations, mse_arr, 'b-')
plt.title("Batch Mode Gradient Descent (Training Data)")
plt.xlabel("iteration count")
plt.ylabel("MSE");

# For verification only
coeffs_m1 = np.polyfit(X_train.flatten(), Y_train, 1, rcond=None, full=False, w=None, cov=False)
print(coeffs_m1)
```

```python
plt.plot(iterations, mse_arr_test, 'r-')
plt.title("Batch Mode Gradient Descent (Testing Data)")
plt.xlabel("iteration count")
plt.ylabel("MSE");
```

```python
w, iter_cnt, mse_arr, mse_arr_test = GradDes(mode=2, w_init=w_init, Xtrain=X_train, Ytrain=Y_train, Xtest=X_test, 
                                     Ytest=Y_test, poly_deg=1, delta=err_tol, step_size=step_sz, max_iter=max_iter)
print(f"learned model: y = {float(w[0])} + {float(w[1])}x")
print(f"iteration count: {iter_cnt}")
print(f"final training MSE: {mse_arr[-1]}")
print(f"final testing MSE: {mse_arr_test[-1]}")

iterations = np.arange(0, iter_cnt+1, 1)

plt.plot(iterations, mse_arr, 'b-')
plt.title("Stochastic Mode Gradient Descent (Training Data)")
plt.xlabel("iteration count")
plt.ylabel("MSE");
```

```python
plt.plot(iterations, mse_arr_test, 'r-')
plt.title("Stochastic Mode Gradient Descent (Testing Data)")
plt.xlabel("iteration count")
plt.ylabel("MSE");
```

#### Assumptions: 
The parameters chosen for this part were the following:
    
    Initial guess vector: zero array
    arbitrary error tolerance for termination: 9.67
    learning rate: 0.01
    maximum iteration count: 5,000

Parameters mentioned above are purely arbitrary and does not necesarily indicate the model's closeness to the actual solution. Values were chosen for the sake of comparison between the two modes of gradient descent namely, batch mode and stochastic mode.

#### Discussion:

With the given constraints as detailed above, it is shown that in terms of accuracy in the testing set, the model obtained via batch mode is far superior than that of the one obtained via stochastic mode. Also, suppose the measure of the convergence speed is purely equal to the number to total iterations until convegrence, the batch mode took much lesser iterations than the stochastic mode. This means that in order to arrive at competitive level of accuracy, the stochastic mode needs to have much more iterations. This behavior is expected since the mode is naturally probabilistic. That means that arriving at much better accuracy using stochastic mode is much more of a by-chance than predictable, and this is due to random gradient sampling.  

With that, it is safe to say that if one is looking for a more accurate and faster algorithm, batch mode is a more practical option. It must be noted, however, that batch mode is prone to issues of local minima where the algorithm fails to converge. This issue is not a problem with the stochastic mode as the gradient continuously jumps from one random sample to another, minimizing the chance of being stuck. So, there is no one-size-fits-all solution to a problem as it purely depends on the target application.

#### b) Plot MSE of the testing set vs. learning rate (using 0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 0.01) and determine the best learning rate.

```python
# Note: Takes around a minute and a half to finish all iterations 

# Function input parameters
w_init = np.array([0, 0])  # initial guess array
degree = 1 # polynomial degree for the model
err_tol = 9.67  # arbitrary error tolerance for termination criterion
step_sz = [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 0.01]  # array of learning rates
max_iter = 5000 # desired max iteration count

# For plotting
mse_test_batch = []
mse_test_stochastic = []

for i in step_sz:
    w, iter_cnt, mse_arr, mse_arr_test = GradDes(mode=1, w_init=w_init, Xtrain=X_train, Ytrain=Y_train, Xtest=X_test, 
                                     Ytest=Y_test, poly_deg=1, delta=err_tol, step_size=i, max_iter=max_iter)

    mse_test_batch.append(mse_arr_test[-1])
    
for i in step_sz:
    w, iter_cnt, mse_arr, mse_arr_test = GradDes(mode=2, w_init=w_init, Xtrain=X_train, Ytrain=Y_train, Xtest=X_test, 
                                     Ytest=Y_test, poly_deg=1, delta=err_tol, step_size=i, max_iter=max_iter)

    mse_test_stochastic.append(mse_arr_test[-1])
    
print(f"Batch Mode MSE values in ascending step size order: {mse_test_batch}")
print(f"Stochastic Mode MSE values in ascending step size order: {mse_test_stochastic}")

plt.plot(step_sz, mse_test_batch, 'b-', label='batch mode')
plt.plot(step_sz, mse_test_stochastic, 'g-', label='stochastic mode')
plt.title("MSE vs Learning Rate (Testing Data Set)")
plt.xlabel("learning rate")
plt.ylabel("MSE")
plt.legend(frameon=True, shadow=True, borderpad=2);
```

#### Discussion:
As seen from the plots and the printed outputs, the best learning rate for the batch mode given the above arbitrary parameter values is 0.007, while the best one for stochastic mode is 0.004. It is again imperative to note that this outcome may vary depending on the constraint values that the user has defined, specifically for the error tolerance and the maximum iteration count. The more we let the algorithms evolve, the greater accuracy we can obtain, and the best learning rate may differ dramatically especially for the stochastic mode whose chraceristic is purely probabilistic.
